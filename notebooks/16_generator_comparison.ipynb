```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
plt.style.use('dark_background')
COLORS = ['#00d4ff', '#ff007f']
METRICS_FILE = "../evaluation/reports/generator_metrics.json"
```

## 1. Load Metric Comparison

```python
with open(METRICS_FILE, 'r') as f:
    results = json.load(f)

df_results = pd.DataFrame(results).T
df_results
```

## 2. Visual Performance Comparison

```python
df_plot = df_results[["avg_rougeL", "avg_bert_score", "avg_structure", "avg_grounding"]]
df_plot.plot(kind='bar', figsize=(12, 6), color=['#00d4ff', '#ff007f', '#39ff14', '#ffd700'])
plt.title("Model Performance Comparison", fontsize=16)
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.grid(alpha=0.1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

## 3. Conclusion & Model Choice

```python
winner = df_results['composite'].idxmax()
print(f"The champion model is: {winner} with a composite score of {df_results.loc[winner, 'composite']:.4f}")
```
